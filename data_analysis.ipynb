```python

import pandas as pd

df= pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'] )

df.head(5)#20 records to see (head)
```

```python
df['Ship Mode'].unique()# we want not available and unknown also in nun so we will add these two in na_valuesin above cell
```

```python
df['Ship Mode'].unique()
```

```python
#Now we will convert column name to lower case and remove spaces in some column name as data should be consistent
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(2)
```

```python
# now we will add columns discount,selling prics and profit
df['discount']=(df['list_price']*df['discount_percent'])*0.01
df['selling_price']=df['list_price']-df['discount']
df['profit']=df['selling_price']-df['cost_price']
df
```

```python
df.dtypes
#We need to convert data type of date to date type
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")

df.drop(columns=['list_price','discount_percent','cost_price'],inplace=True)
#inplace is used so that drop of columns reflect on our database too otherewise its temperory droped
df
```
